```python
Q1 = df['total_bill'].quantile(0.25)
Q1
```

```python
IQR = Q3 - Q1
IQR
```

```python
upper = Q3 + (1.5*IQR)
upper
```

```python
lower = Q1 - (1.5*IQR)
lower
```
